```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
w = 0.076
N = 10
k_s = 0.805
tau = 2*np.pi /w

t = np.linspace(0,tau,1000)
envelope_sin = np.sin(np.pi*t/tau)**2
envelope_gauss = np.exp(-1*np.log(2)*np.power(2 *(t-tau/2)/tau, 2.0))
I = (2.0e14/3.15e16) * envelope_sin

U_p = I/(4*w**2)

n = (k_s**2 /2 + 0.5 + U_p)/w


```

```python
# Find integer-crossing points
integer_indices = np.where(np.diff(np.floor(n)))[0]

# Plot curve
plt.plot(t,n)

for idx in integer_indices:
    x_int = t[idx]
    y_int = n[idx]
    plt.plot([x_int, x_int], [np.min(n), y_int], color='k')
    plt.text(x_int, y_int + 0.25, ha='center',s = f"{round(n[idx])}")
plt.show()

```
